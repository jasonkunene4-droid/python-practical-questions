10. What is the difference between implicit and explicit type casting?
	Implicit type casting is when python automatically performs the conversion of the data type for example adding a float and an integer can automatically give us a float. While explicit type casting requires the programmer to convert the data types manually for example, we cannot add a string and a float unless one is converted to the other, in short, we cannot concatenate theses two which changing them into a singular data type.

```python
14. #reversing a string using a while loop
word="jason"
print(word)

reversed_word=('')
i=len(word)-1

while i>=0:
    reversed_word=reversed_word+word[i]
    i=i-1

print("reversed word:",reversed_word)
```

```python
15. # program to calculate the factorial of a given number
num = int(input("Enter a number to find its factorial: "))

factorial = 1

while num > 1:
    factorial *= num
    num -= 1

print("Factorial is:", factorial)

```
